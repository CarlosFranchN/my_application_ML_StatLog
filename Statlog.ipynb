```python
def IQR(df,colu):
  col = df[colu]
  Q1 = df[colu].quantile(.25)
  Q3 = df[colu].quantile(.75)
  IQR = Q3 - Q1

  limite_inferior = Q1 - 1.5 * IQR
  limite_superior = Q3 + 1.5 * IQR
  # outliers = df[colu][(df[colu] > limite_inferior) & (df[colu] < limite_superior)]
  return limite_inferior, limite_superior
```

```python
def retirando_outliers(df,col):
  lim_inf,lim_sup = IQR(df,col)
  df = df[(df[col] > lim_inf) & (df[col] < lim_sup)]
  return df
```

```python
def plotar_precision_recall_vs_threshold(precisions,recalls,thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision", linewidth=2)
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall", linewidth=2)
    plt.legend(loc="center right", fontsize=16) # Not shown in the book
    plt.xlabel("Threshold", fontsize=16)        # Not shown
    plt.grid(True)                              # Not shown
    plt.axis([-50000, 50000, 0, 1])             # Not shown
```

```python
def plotar_precision_vs_recall(precisions, recalls):
    plt.figure(figsize=(8, 6))  # Define o tamanho da figura

    # Plota a curva Precision-Recall
    sns.lineplot(x=recalls, y=precisions, label="Precision vs Recall", color="b", linewidth=2)

    # Ajuste de rótulos e título
    plt.xlabel("Recall", fontsize=14)
    plt.ylabel("Precision", fontsize=14)
    plt.title("Curva Precision-Recall", fontsize=16)

    # Ajustando os limites do gráfico
    plt.xlim([0, 1])
    plt.ylim([0, 1])

    # Melhorando a grade
    plt.grid(visible=True, linestyle="--", alpha=0.6)

    # Exibir legenda corretamente
    plt.legend(loc="best", fontsize=12)

    # Mostrar o gráfico
    plt.show()
```

```python
models = {
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Logistic Regression": LogisticRegression(),
    "Support Vector Machine": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
}
```

```python
for name, model in models.items():
    # Treina com dados balanceados
    model.fit(X_train_resampled, y_train_resampled)

    # Preve nos dados de teste REAIS (desbalanceados e nunca vistos)
    y_pred = model.predict(X_test_scaled)

    # Métricas
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted') # weighted é bom p/ classes desbalanceadas no teste

    print(f"--- {name} ---")
    print(classification_report(y_test, y_pred))

    results[name] = {"Accuracy": acc, "F1-score": f1}
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
resultados_cv = {}
```

```python
for name, model in models.items():
    f1_scores = []

    
    for train_index, val_index in cv.split(X_train_scaled, y_train):

        
        X_train_fold, X_val_fold = X_train_scaled[train_index], X_train_scaled[val_index]
        y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]

        
        ros_fold = RandomOverSampler(random_state=42)
        X_train_fold_res, y_train_fold_res = ros_fold.fit_resample(X_train_fold, y_train_fold)

        
        model.fit(X_train_fold_res, y_train_fold_res)

        
        y_pred_fold = model.predict(X_val_fold)
        f1_scores.append(f1_score(y_val_fold, y_pred_fold, average='weighted'))

    mean_score = np.mean(f1_scores)
    resultados_cv[name] = mean_score
    print(f"{name}: Média F1-Score (Correct CV): {mean_score:.4f}")
```

```python
import os


lucro_cliente = 1000
prejuizo_calote = -5000
model_names = ['Decision Tree', 'Random Forest', 'Logistic Regression', 'K-Nearest Neighbors']
thresholds = np.arange(0.0, 1.01, 0.05)


plt.figure(figsize=(12, 8))


melhor_saldo_geral = -float('inf')
melhor_config_geral = ""


for name in model_names:
    model = best_model[name]
    if hasattr(model, "predict_proba"):
        probs_mau = model.predict_proba(X_test)[:, 1]
    else:
        continue
    
   
    probs_mau = model.predict_proba(X_test)[:, 1]
    
    lucros_modelo = []
    
    for t in thresholds:
        
        y_pred_t = np.where(probs_mau > t, 2, 1)
        
        
        cm = confusion_matrix(y_test, y_pred_t)
        tn, fp, fn, tp = cm.ravel()
        
        
        saldo = (tn * lucro_cliente) + (fn * prejuizo_calote)
        lucros_modelo.append(saldo)
        
        
        if saldo > melhor_saldo_geral:
            melhor_saldo_geral = saldo
            melhor_config_geral = f"{name} (Corte > {t:.2f})"

    
    plt.plot(thresholds, lucros_modelo, label=name, marker='.', linewidth=2)


plt.axhline(y=0, color='r', linestyle='--', label='Zero a Zero (Break-even)', linewidth=2)
plt.title('Comparação de Lucratividade: Modelos e Limiares de Risco', fontsize=14)
plt.xlabel('Limite de Risco Aceitável (Threshold)', fontsize=12)
plt.ylabel('Lucro da Carteira (R$)', fontsize=12)
plt.legend(loc='lower left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlim(0, 1.0)

pasta_img = 'img'
os.makedirs(pasta_img, exist_ok=True)
caminho_arquivo = os.path.join(pasta_img, 'analise_lucratividade.png')
plt.savefig(caminho_arquivo, dpi=300, bbox_inches='tight')

plt.show()

print("="*60)
print(f"🏆 ESTRATÉGIA CAMPEÃ ABSOLUTA: {melhor_config_geral}")
print(f"💰 LUCRO MÁXIMO POSSÍVEL: R$ {melhor_saldo_geral:,.2f}")
print("="*60)
```
